#modèle pré-configuré haarcascade_frontalface_default.xml

```python
# on réimporte UNE image, détecte l'image, transforme en gris, puis le .xml sert à détecter le VISAGE
image = cv2.imread("C:/Users/ahwid/Desktop/Licence Pro Big Data/PROJET TUT TONIC FORME/Capture_v2/1.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #en gris

dirCascadeFiles = "C:/Users/ahwid/Desktop/Licence Pro Big Data/PROJET TUT TONIC FORME/reconnaissance-faciale-master-/"
# Load HaarCascade from the file with OpenCV
faceCascade = dirCascadeFiles + "haarcascade_frontalface_default.xml"
eye_cascade = dirCascadeFiles + "haarcascade_eye.xml"
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier("haarcascade_eye.xml")
```

```python
#boucle sur l'ensemble des photos
listeImages_v2 = []
for i in range(1, len(listeImages)):
    image = cv2.imread("C:/Users/ahwid/Desktop/Licence Pro Big Data/PROJET TUT TONIC FORME/Capture_v2/"+listeImages[i])
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    plt.show()
    
    faces = face_cascade.detectMultiScale(
        image,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30,30),
        flags = cv2.CASCADE_SCALE_IMAGE
    )
    
    im = image.copy()
    
    for (x, y, w, h) in faces:
        cv2.rectangle(im, (x,y), (x+w, y+h), (0, 255, 0), 2)
        (x, y, w, h) = faces[0]
        center_x = x+w/2
        center_y = y+h/2
        
        crop_img = im[y:y+h, x:x+w]
        
        listeImages_v2.append(crop_img)
    
                       
    plt.imshow(crop_img, cmap = 'gray')
```
